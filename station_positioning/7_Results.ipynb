```python
import pandas as pd
import yaml 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from pylab import *

import seaborn as sns

with open('../conf.yaml') as f:
    conf = yaml.load(f, Loader=yaml.FullLoader)

pathGeneral = conf["pathGeneral"]
pathGeneral
```

# Overall results
- 18 months <p>
and 
- 3 months high (2008_11_1-2009_1_31)
- 3 months low (2009_9_1-2009_11_30)

```python
sns.set(style="ticks",font_scale=1.4)
res = pd.read_csv(pathGeneral+"results/results.txt")
res = res[res["period"]=="2008_7_1-2009_12_31"]
fontsize = 10

sns.catplot(x="strategy",y="ratio",hue="stations",col="region", row="period", data=res, kind="bar",ci = None,aspect=1.3,margin_titles=True)

savefig(pathGeneral+"results/ratio_covered_period1.png", dpi  = 300)

sns.set(style="ticks",font_scale=1.6)
res = pd.read_csv(pathGeneral+"results/results.txt")
res = res[(res["period"]=="2008_11_1-2009_1_31") | (res["period"]=="2009_9_1-2009_11_30")]
fontsize = 10

sns.catplot(x="strategy",y="ratio",hue="stations",col="region", row="period", data=res, kind="bar",ci = None,aspect=1.3,margin_titles=True)

savefig(pathGeneral+"results/ratio_covered_period2_3.png", dpi  = 300)
```
